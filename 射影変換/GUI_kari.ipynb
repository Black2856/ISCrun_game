```python
import tkinter as tk

class Application(tk.Frame):
    def __init__(self, master = None):
        super().__init__(master)
        self.pack()
        
        self.master.title("タイトル")    # ウィンドウタイトル
        self.master.geometry("1000x700") # ウィンドウサイズ(幅x高さ)

        # Canvasの作成
        self.canvas1 = tk.Canvas(
            self.master, 
            width = 510,
            height = 400,
            bg = "cyan"
            )
        self.canvas2 = tk.Canvas(
            self.master, 
            width = 250,
            height = 200,
            bg = "yellow"
            )
        self.canvas3 = tk.Canvas(
            self.master, 
            width = 250,
            height = 200,
            bg = "pink"
            )
        

        # labelを作成
        self.img_label1 = tk.Label(text="処理サイズ")
        self.img_label2 = tk.Label(text="×")
        # textboxを作成
        self.tex_WIDTH = tk.Entry(self.master,
                             width = 4,
                             )
        
        self.tex_HEIGHT = tk.Entry(self.master,
                             width = 4,
                             )
                             
        # スライサーを作成
        self.scale_HSV_H = tk.Scale(self.master,
                              orient = tk.HORIZONTAL,
                              length = 350,
                              width = 15,
                               label = "色サークル検出H",
                               sliderlength = 20,
                               to = 255,
                               tickinterval=25)
        
        self.scale_HSV_S = tk.Scale(self.master,
                              orient = tk.HORIZONTAL,
                              length = 350,
                              width = 15,
                               label = "色サークル検出S",
                               sliderlength = 20,
                               to = 255,
                               tickinterval=25)
        
        self.scale_HSV_V = tk.Scale(self.master,
                              orient = tk.HORIZONTAL,
                              length = 350,
                              width = 15,
                               label = "色サークル検出V",
                               sliderlength = 20,
                               to = 255,
                               tickinterval=25)
        
        self.scale_LINE_BLACK = tk.Scale(self.master,
                              orient = tk.HORIZONTAL,
                              length = 350,
                              width = 15,
                               label = "黒ライン検出V",
                               sliderlength = 20,
                               to = 255,
                               tickinterval=25)
        
        self.scale_LINE_stringent = tk.Scale(self.master,
                              orient = tk.HORIZONTAL,
                              length = 350,
                              width = 15,
                               label = "ライン検知の厳しさ",
                               sliderlength = 20,
                               to = 1000,
                               tickinterval=100)
        
        self.scale_LINE_angle = tk.Scale(self.master,
                              orient = tk.HORIZONTAL,
                              length = 350,
                              width = 15,
                               label = "ライン検知の角度",
                               sliderlength = 20,
                               resolution = 0.1,
                               to = 3,
                               tickinterval=0.5)
        
        self.scale_point_max = tk.Scale(self.master,
                              orient = tk.HORIZONTAL,
                              length = 350,
                              width = 15,
                               label = "点検出の最大",
                               sliderlength = 20,
                               to = 1000,
                               tickinterval=100)
        
        self.scale_point_min = tk.Scale(self.master,
                              orient = tk.HORIZONTAL,
                              length = 350,
                              width = 15,
                               label = "点検出の最小",
                               sliderlength = 20,
                               to = 1000,
                               tickinterval=100)
        
        
        
        # 配置
        self.canvas1.place(x=20, y=20)
        self.canvas2.place(x=20, y=440)
        self.canvas3.place(x=280, y=440)        
        self.scale_HSV_H.place(x=600, y=70)
        self.scale_HSV_S.place(x=600, y=140)
        self.scale_HSV_V.place(x=600, y=210)
        self.scale_LINE_BLACK.place(x=600, y=280)
        self.scale_LINE_stringent.place(x=600,y=350)
        self.scale_LINE_angle.place(x=600,y=420)
        self.scale_point_min.place(x=600,y=490)
        self.scale_point_max.place(x=600,y=560)
        self.tex_WIDTH.place(x=700,y=20)
        self.tex_HEIGHT.place(x=870,y=20)
        self.img_label1.place(x=600,y=20)
        self.img_label2.place(x=800,y=20)
        
if __name__ == "__main__":
    root = tk.Tk()
    app = Application(master = root)
    app.mainloop()

        
```
